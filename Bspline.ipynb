### B-Spline

```python
# https://en.wikipedia.org/wiki/B-spline
# https://www.cs.unc.edu/~dm/UNC/COMP258/LECTURES/B-spline.pdf
# https://personal.math.vt.edu/embree/math5466/lecture10.pdf
```

![Definition](./Bspline/bSplineDef.jpg)

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

#plotting parameters
msize = 3
```

```python
# Knot Array
i_len = 6 # i length 
t_max = i_len + 1 # t goes from 0 to this value
t_res = 201 # resolution 

# init 
t = np.linspace(0, t_max, t_res) 
Knot_arr = np.zeros((i_len,len(t)))

# add Uniform Knot Sequence
for y in range(i_len):
    for z in range(len(t)):
        if t[z] >= y and t[z] < y+1: 
            Knot_arr[y,z] = 1

```

### Example below of Uniform Knot Sequence

![UKS](./Bspline/UniformKnotSequence.jpg)

```python
# plot N_i,1 for a  Uniform Knot Sequence
fig, axs = plt.subplots(i_len, 1, layout='tight')
for z in range(i_len):
    axs[z].plot(t,Knot_arr[z,:], marker='o', linestyle='', color='k',markersize=msize)
    axs[z].grid(True)
    axs[z].set_ylim(-0.5,1.5)
```

```python
# B-Spline Curve k = 1
# Sum over all i
B1 = np.sum(Knot_arr,0)
```

```python
plt.figure(figsize=(7,1))
plt.plot(t,B1, marker='o', linestyle='', color='k',markersize=msize)
plt.grid(True)
plt.ylim(-0.5,1.5)
```

### Now for the more interesting k = 2; Known as Hat (linear interpolation)

![nik](./Bspline/nik.jpg)

![n02](./Bspline/n02.jpg)

![n12](./Bspline/n12.jpg)

```python
# Init for k=2
k = 2
N2 = np.zeros((i_len-k+1,len(t)))
for x in range(i_len-k+1):
    N2[x,:] = ((t-x)/(k-1))*Knot_arr[x,:]+((k+x-t)/(k-1))*Knot_arr[x+1,:]
```

```python
# plot N_i,1 for a  Uniform Knot Sequence
fig, axs = plt.subplots(i_len-k+1, 1, layout='tight')
for z in range(i_len-k+1):
    axs[z].plot(t,N2[z,:], marker='o', linestyle='', color='k',markersize=msize)
    axs[z].grid(True)
    axs[z].set_ylim(-0.5,1.5)
```

```python
# B-Spline Curve k = 2
# Sum over all i
B2 = np.sum(N2,0)
```

```python
plt.figure(figsize=(7,1))
plt.plot(t,B2, marker='o', linestyle='', color='k',markersize=msize)
plt.grid(True)
plt.ylim(-0.5,1.5)
```

### Compare different orders

```python
# plot N_i,1 for a  Uniform Knot Sequence
k=2
colors = cm.plasma(np.linspace(0.2, 0.8, i_len-k+1))

# Remove zeros for plot
N2_plt = N2
N2_plt[N2_plt==0] = np.nan
B2_plt = B2
B2_plt[B2_plt==0] = np.nan


plt.figure(figsize=(7,3))
for z in range(i_len-k+1):
    plt.plot(t,N2_plt[z,:], marker='', linestyle='--', color=colors[z],linewidth = 2, markersize=1)
plt.plot(t,B2, marker='', linestyle='--', color='k',linewidth = 1, markersize=1)
plt.ylim(-0.2,1.2)
plt.grid(alpha=0.4)  
plt.show()
```
