```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
# Just getting a feeling for odds as a function of sequential problalities
# important for gate fidelity in Quantum Computing, Yield in Chemistry, Coin Flips... 
```

```python
#init arrays
prob_arr_perc =  np.array([90,91,92,93,94,95,95,96,97,98,99,99.9,99.99,99.999,99.9999])
prob_arr = 0.01 * prob_arr_perc
err_arr = 1 - prob_arr
counts = np.arange(1, 1000000, 1)
Odds = np.zeros((len(prob_arr),len(counts))) 
```

```python
# Calc Odds
i=0
j=0
for prob in prob_arr:
    j=0
    for count in counts:
        Odds[i,j] = prob**count
        j+=1
    i+=1
    
```

```python
colors = cm.plasma(np.linspace(0, 0.9, Odds.shape[0]))

# Plot each row as a separate line with colormap
plt.figure(figsize=(8, 5))
for i, (row, color) in enumerate(zip(Odds, colors)):
    plt.plot(counts, row, label=f'Row {prob_arr_perc[i]}', color=color, linewidth=2)

# Add grey box in the background between y=0.5 and y=0
plt.axhspan(0, 0.5, color='grey', alpha=0.3)  # alpha controls transparency
    
plt.xlabel("Count")
plt.ylabel("Probability")
plt.title("Odds of Successive Sequential Probability Events")
plt.xlim(1,10)
plt.ylim(0,1.05)
plt.grid(alpha=0.4)  
# plt.legend()
plt.show()
```

```python
# Check when drops below 0.5, aka 50/50 chance, threshold for computing
threshold = 0.5
dip_indices = np.argmax(Odds < threshold, axis=1)
# If a row never dips below the threshold, set to -1
dip_indices[np.all(Odds >= threshold, axis=1)] = -1

print("Starting Probability: ", prob_arr_perc, "%\n")
print("Number of events before probability drops below 0.5: ", dip_indices)
```
